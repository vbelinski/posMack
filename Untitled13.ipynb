Proposta: Concatenar os textos de Alexandre Garcia e Arnaldo Jabor, aplicar as tecnicas de classificação de texto para descobrir o autor da frase.

```python
import pandas as pd
df = pd.read_json('https://raw.githubusercontent.com/lucianomcsilva/articulistas/main/alexandre_garcia.json')
df1 = pd.read_json('https://raw.githubusercontent.com/lucianomcsilva/articulistas/main/arnaldo_jabor.json')
```

```python
df.sample(10)
```

```python
df1.sample(10)
```

```python
#Amostra igual para todas as categorias
df_final2 = df.sample(80)
df_final3 = df1.sample(80)
```

```python
df_final = pd.concat([df_final2, df_final3]).sample(frac=1)
#df_final.groupby(['category']).count()
```
